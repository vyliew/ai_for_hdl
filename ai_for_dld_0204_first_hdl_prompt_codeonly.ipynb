# AND, OR, NOT Logic – Verilog and VHDL Simulation

This notebook provides a step-by-step guide to generating and simulating a fundamental logic module in both Verilog and VHDL. We will create a simple circuit combining AND, OR, and NOT gates, then simulate its behavior and visualize the results.

### What Our Logic Module Will Do
Our module will take two single-bit inputs, `a` and `b`. It will produce three single-bit outputs:
*   `out_and`: The logical AND of `a` and `b`.
*   `out_or`: The logical OR of `a` and `b`.
*   `out_not`: The logical NOT of `a`.

### Why Both Verilog and VHDL?

*   **Flexibility:** Depending on the project or team, you might need to adapt to either language.

### Tools We'll Use
We will leverage powerful open-source tools:
*   **GHDL:** A VHDL compiler and simulator. It translates VHDL code into an executable for simulation.
*   **Icarus Verilog (iverilog):** A Verilog compiler and simulator. It compiles Verilog code into an executable for simulation.
*   **Python:** Used within Google Colab for orchestrating the workflow, installing tools, writing files, and managing simulation outputs.

### How Simulation Outputs Are Generated
After creating our logic module and testbenches in both HDLs, we'll run simulations. These simulations will:
1.  Apply a sequence of input values to our logic module (defined in testbenches).
2.  Record the changes in input and output signals over time.
3.  Generate waveform files (specifically, Value Change Dump - VCD files). These `.vcd` files are a standard format that can be viewed with waveform viewers to visually inspect the circuit's behavior.

### 🔹 Verilog Version
This Verilog module takes two single-bit inputs, `a` and `b`, and produces three single-bit outputs. It calculates the logical AND of `a` and `b`, the logical OR of `a` and `b`, and the logical NOT of `a`.

```python
%%writefile and_or_not.v
// Define a module named and_or_not
module and_or_not (
    input a,  // Declare input 'a'
    input b,  // Declare input 'b'
    output out_and, // Declare output for AND operation
    output out_or,  // Declare output for OR operation
    output out_not  // Declare output for NOT operation
);

    // Implement the AND gate: out_and is true if both a and b are true
    assign out_and = a & b;

    // Implement the OR gate: out_or is true if either a or b is true (or both)
    assign out_or = a | b;

    // Implement the NOT gate: out_not is true if a is false
    assign out_not = ~a;

endmodule // End of the and_or_not module
```

### 🔹 VHDL Version
This VHDL entity implements the same logic as the Verilog version. It defines an entity with two single-bit inputs, `a` and `b`, and three single-bit outputs, `out_and`, `out_or`, and `out_not`. The architecture section describes the behavior of these logic gates.

```python
%%writefile and_or_not.vhdl
-- Import the standard logic library
library ieee;
use ieee.std_logic_1164.all;

-- Declare the entity named and_or_not
entity and_or_not is
    port (
        a       : in  std_logic;  -- Declare input 'a' of type std_logic
        b       : in  std_logic;  -- Declare input 'b' of type std_logic
        out_and : out std_logic;  -- Declare output for AND operation
        out_or  : out std_logic;  -- Declare output for OR operation
        out_not : out std_logic   -- Declare output for NOT operation
    );
end entity and_or_not;

-- Define the architecture named behavioral for the and_or_not entity
architecture behavioral of and_or_not is
begin
    -- Implement the AND gate: out_and is '1' if both a and b are '1'
    out_and <= a and b;

    -- Implement the OR gate: out_or is '1' if either a or b is '1' (or both)
    out_or <= a or b;

    -- Implement the NOT gate: out_not is '1' if a is '0'
    out_not <= not a;

end architecture behavioral; -- End of the behavioral architecture
```

### 🔹 Verilog Testbench
This Verilog testbench will verify the functionality of the `and_or_not` module. It applies all four possible input combinations for `a` and `b` (00, 01, 10, 11) with a delay between each change. It also sets up waveform dumping to a VCD file (`verilog_output.vcd`) so you can visualize the signals over time.

```python
%%writefile and_or_not_tb.v
// Define a testbench module
module and_or_not_tb;

    // Declare signals for inputs and outputs
    reg a;  // Testbench input for 'a'
    reg b;  // Testbench input for 'b'
    wire out_and; // Testbench output for AND
    wire out_or;  // Testbench output for OR
    wire out_not; // Testbench output for NOT

    // Instantiate the Unit Under Test (UUT) - our and_or_not module
    and_or_not dut (
        .a(a),
        .b(b),
        .out_and(out_and),
        .out_or(out_or),
        .out_not(out_not)
    );

    // Initial block for stimulus generation
    initial begin
        // Open VCD file for waveform dumping
        $dumpfile("verilog_output.vcd");
        // Dump all signals in the current scope
        $dumpvars(0, and_or_not_tb);

        // Initialize inputs
        a = 0; b = 0;
        #10; // Wait for 10 time units

        // Test case 1: a=0, b=1
        a = 0; b = 1;
        #10;

        // Test case 2: a=1, b=0
        a = 1; b = 0;
        #10;

        // Test case 3: a=1, b=1
        a = 1; b = 1;
        #10;

        // End simulation
        $finish;
    end

endmodule // End of testbench module
```

### 🔹 VHDL Testbench
This VHDL testbench simulates the `and_or_not` entity by applying a sequence of inputs for `a` and `b` (00, 01, 10, 11) with time delays between each input change. It utilizes GHDL-specific commands to generate a VCD waveform file (`vhdl_output.vcd`) to help visualize the signal behavior during simulation.

```python
%%writefile and_or_not_tb.vhdl
-- Import standard VHDL libraries
library ieee;
use ieee.std_logic_1164.all;

-- Import GHDL-specific package for VCD waveform dumping
-- This package is an extension provided by GHDL for VCD generation.
library GHDL;
use GHDL.vcd.all;

-- Testbench entity declaration
entity and_or_not_tb is
end entity and_or_not_tb;

-- Testbench architecture definition
architecture test_bench of and_or_not_tb is

    -- Component declaration for the Unit Under Test (UUT)
    component and_or_not
        port (
            a       : in  std_logic;
            b       : in  std_logic;
            out_and : out std_logic;
            out_or  : out std_logic;
            out_not : out std_logic
        );
    end component and_or_not;

    -- Declare signals to connect to the UUT's ports
    signal a_s       : std_logic := '0'; -- Testbench input signal for 'a', initialized to '0'
    signal b_s       : std_logic := '0'; -- Testbench input signal for 'b', initialized to '0'
    signal out_and_s : std_logic;      -- Testbench output signal for AND
    signal out_or_s  : std_logic;      -- Testbench output signal for OR
    signal out_not_s : std_logic;      -- Testbench output signal for NOT

begin

    -- Instantiate the Unit Under Test (UUT)
    UUT: and_or_not
        port map (
            a       => a_s,
            b       => b_s,
            out_and => out_and_s,
            out_or  => out_or_s,
            out_not => out_not_s
        );

    -- Stimulus process to apply input patterns and generate VCD
    stimulus_process: process
    begin
        -- Set the VCD file name
        GHDL_DUMPFILER("vhdl_output.vcd");
        -- Dump all signals in the current scope for waveform visualization
        GHDL_DUMPVARS(test_bench);

        -- Initial inputs: a=0, b=0
        a_s <= '0';
        b_s <= '0';
        wait for 10 ns; -- Wait for 10 nanoseconds

        -- Test case 1: a=0, b=1
        a_s <= '0';
        b_s <= '1';
        wait for 10 ns;

        -- Test case 2: a=1, b=0
        a_s <= '1';
        b_s <= '0';
        wait for 10 ns;

        -- Test case 3: a=1, b=1
        a_s <= '1';
        b_s <= '1';
        wait for 10 ns;

        -- End simulation by pausing indefinitely
        wait;
    end process stimulus_process;

end architecture test_bench;
```

### 🔸 Verilog Simulation
This block will compile your Verilog module and testbench using `iverilog`, run the simulation with `vvp`, and then organize the generated output files, including the `.vcd` waveform file, into a dedicated `verilog_outputs/` folder for easy access.

```python
%%bash
# Create a directory to store Verilog outputs
mkdir -p verilog_outputs/

# Compile Verilog design and testbench
# -o and_or_not_vvp: Specifies the output executable file name
# and_or_not.v: Your main Verilog module file
# and_or_not_tb.v: Your Verilog testbench file
iverilog -o and_or_not_vvp and_or_not.v and_or_not_tb.v

# Run the compiled Verilog simulation
# This will execute the testbench and generate verilog_output.vcd
vvp and_or_not_vvp

# Move generated files into the output directory
mv and_or_not_vvp verilog_outputs/
mv verilog_output.vcd verilog_outputs/

echo "✅ Verilog simulation complete. Outputs are in verilog_outputs/"
```

### 🔸 VHDL Simulation
This block will analyze and elaborate your VHDL module and testbench using `ghdl`, then run the simulation. It will generate a `.vcd` waveform file and store all simulation-related outputs in a `vhdl_outputs/` folder, ensuring a clean workspace.

```python
%%bash
# Create a directory to store VHDL outputs
mkdir -p vhdl_outputs/

# Analyze VHDL design and testbench files
# ghdl -a: Analyzes the VHDL source files
ghdl -a and_or_not.vhdl and_or_not_tb.vhdl

# Elaborate the testbench
# ghdl -e: Elaborates the design unit (the testbench in this case)
ghdl -e and_or_not_tb

# Run the VHDL simulation
# ghdl -r: Runs the elaborated design unit
# --vcd=vhdl_output.vcd: Specifies to generate a VCD waveform file
ghdl -r and_or_not_tb --vcd=vhdl_output.vcd

# Move generated files (work-obj93.ghdl, vhdl_output.vcd) into the output directory
mv work-obj93.ghdl vhdl_outputs/
mv vhdl_output.vcd vhdl_outputs/

echo "✅ VHDL simulation complete. Outputs are in vhdl_outputs/"
```

### 🔹 Verilog Version
This Verilog module takes two single-bit inputs, `a` and `b`, and produces three single-bit outputs. It calculates the logical AND of `a` and `b`, the logical OR of `a` and `b`, and the logical NOT of `a`.

```python
%%writefile and_or_not.v
// Define a module named and_or_not
module and_or_not (
    input a,  // Declare input 'a'
    input b,  // Declare input 'b'
    output out_and, // Declare output for AND operation
    output out_or,  // Declare output for OR operation
    output out_not  // Declare output for NOT operation
);

    // Implement the AND gate: out_and is true if both a and b are true
    assign out_and = a & b;

    // Implement the OR gate: out_or is true if either a or b is true (or both)
    assign out_or = a | b;

    // Implement the NOT gate: out_not is true if a is false
    assign out_not = ~a;

endmodule // End of the and_or_not module
```

### 🔹 VHDL Version
This VHDL entity implements the same logic as the Verilog version. It defines an entity with two single-bit inputs, `a` and `b`, and three single-bit outputs, `out_and`, `out_or`, and `out_not`. The architecture section describes the behavior of these logic gates.

### 🔹 Verilog Testbench
This Verilog testbench will verify the functionality of the `and_or_not` module. It applies all four possible input combinations for `a` and `b` (00, 01, 10, 11) with a delay between each change. It also sets up waveform dumping to a VCD file (`verilog_output.vcd`) so you can visualize the signals over time.

```python
%%writefile and_or_not_tb.v
// Define a testbench module
module and_or_not_tb;

    // Declare signals for inputs and outputs
    reg a;  // Testbench input for 'a'
    reg b;  // Testbench input for 'b'
    wire out_and; // Testbench output for AND
    wire out_or;  // Testbench output for OR
    wire out_not; // Testbench output for NOT

    // Instantiate the Unit Under Test (UUT) - our and_or_not module
    and_or_not dut (
        .a(a),
        .b(b),
        .out_and(out_and),
        .out_or(out_or),
        .out_not(out_not)
    );

    // Initial block for stimulus generation
    initial begin
        // Open VCD file for waveform dumping
        $dumpfile("verilog_output.vcd");
        // Dump all signals in the current scope
        $dumpvars(0, and_or_not_tb);

        // Initialize inputs
        a = 0; b = 0;
        #10; // Wait for 10 time units

        // Test case 1: a=0, b=1
        a = 0; b = 1;
        #10;

        // Test case 2: a=1, b=0
        a = 1; b = 0;
        #10;

        // Test case 3: a=1, b=1
        a = 1; b = 1;
        #10;

        // End simulation
        $finish;
    end

endmodule // End of testbench module
```

### 🔹 VHDL Testbench
This VHDL testbench simulates the `and_or_not` entity by applying a sequence of inputs for `a` and `b` (00, 01, 10, 11) with time delays between each input change. It utilizes GHDL-specific commands to generate a VCD waveform file (`vhdl_output.vcd`) to help visualize the signal behavior during simulation.

```python
%%writefile and_or_not_tb.vhdl
-- Import standard VHDL libraries
library ieee;
use ieee.std_logic_1164.all;

-- Import GHDL-specific package for VCD waveform dumping
-- This package is an extension provided by GHDL for VCD generation.
library GHDL;
use GHDL.vcd.all;

-- Testbench entity declaration
entity and_or_not_tb is
end entity and_or_not_tb;

-- Testbench architecture definition
architecture test_bench of and_or_not_tb is

    -- Component declaration for the Unit Under Test (UUT)
    component and_or_not
        port (
            a       : in  std_logic;
            b       : in  std_logic;
            out_and : out std_logic;
            out_or  : out std_logic;
            out_not : out std_logic
        );
    end component and_or_not;

    -- Declare signals to connect to the UUT's ports
    signal a_s       : std_logic := '0'; -- Testbench input signal for 'a', initialized to '0'
    signal b_s       : std_logic := '0'; -- Testbench input signal for 'b', initialized to '0'
    signal out_and_s : std_logic;      -- Testbench output signal for AND
    signal out_or_s  : std_logic;      -- Testbench output signal for OR
    signal out_not_s : std_logic;      -- Testbench output signal for NOT

begin

    -- Instantiate the Unit Under Test (UUT)
    UUT: and_or_not
        port map (
            a       => a_s,
            b       => b_s,
            out_and => out_and_s,
            out_or  => out_or_s,
            out_not => out_not_s
        );

    -- Stimulus process to apply input patterns and generate VCD
    stimulus_process: process
    begin
        -- Set the VCD file name
        GHDL_DUMPFILER("vhdl_output.vcd");
        -- Dump all signals in the current scope for waveform visualization
        GHDL_DUMPVARS(test_bench);

        -- Initial inputs: a=0, b=0
        a_s <= '0';
        b_s <= '0';
        wait for 10 ns; -- Wait for 10 nanoseconds

        -- Test case 1: a=0, b=1
        a_s <= '0';
        b_s <= '1';
        wait for 10 ns;

        -- Test case 2: a=1, b=0
        a_s <= '1';
        b_s <= '0';
        wait for 10 ns;

        -- Test case 3: a=1, b=1
        a_s <= '1';
        b_s <= '1';
        wait for 10 ns;

        -- End simulation by pausing indefinitely
        wait;
    end process stimulus_process;

end architecture test_bench;
```

### 🔸 Verilog Simulation
This block will compile your Verilog module and testbench using `iverilog`, run the simulation with `vvp`, and then organize the generated output files, including the `.vcd` waveform file, into a dedicated `verilog_outputs/` folder for easy access.

```python
%%bash
# Create a directory to store Verilog outputs
mkdir -p verilog_outputs/

# Compile Verilog design and testbench
# -o and_or_not_vvp: Specifies the output executable file name
# and_or_not.v: Your main Verilog module file
# and_or_not_tb.v: Your Verilog testbench file
iverilog -o and_or_not_vvp and_or_not.v and_or_not_tb.v

# Run the compiled Verilog simulation
# This will execute the testbench and generate verilog_output.vcd
vvp and_or_not_vvp

# Move generated files into the output directory
mv and_or_not_vvp verilog_outputs/
mv verilog_output.vcd verilog_outputs/

echo "✅ Verilog simulation complete. Outputs are in verilog_outputs/"
```

### 🔸 VHDL Simulation
This block will analyze and elaborate your VHDL module and testbench using `ghdl`, then run the simulation. It will generate a `.vcd` waveform file and store all simulation-related outputs in a `vhdl_outputs/` folder, ensuring a clean workspace.

```python
%%bash
# Create a directory to store VHDL outputs
mkdir -p vhdl_outputs/

# Analyze VHDL design and testbench files
# ghdl -a: Analyzes the VHDL source files
ghdl -a and_or_not.vhdl and_or_not_tb.vhdl

# Elaborate the testbench
# ghdl -e: Elaborates the design unit (the testbench in this case)
ghdl -e and_or_not_tb

# Run the VHDL simulation
# ghdl -r: Runs the elaborated design unit
# --vcd=vhdl_output.vcd: Specifies to generate a VCD waveform file
ghdl -r and_or_not_tb --vcd=vhdl_output.vcd

# Move generated files (work-obj93.ghdl, vhdl_output.vcd) into the output directory
mv work-obj93.ghdl vhdl_outputs/
mv vhdl_output.vcd vhdl_outputs/

echo "✅ VHDL simulation complete. Outputs are in vhdl_outputs/"
```
